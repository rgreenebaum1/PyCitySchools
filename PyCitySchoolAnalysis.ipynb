```python
#Dependencies
import pandas as pd
import numpy as np
import os
import csv
```

```python
# Files to Load
schools_complete = "schools_complete.csv"
students_complete = "students_complete.csv"

print(schools_complete)
print(students_complete)



```

```python
#create dataframe using mapping

schools_df = pd.read_csv(schools_complete)

students_df = pd.read_csv(students_complete)


#read files
schools_df.head()

students_df.head()

```

```python
#Merge


merge_df = pd.merge(schools_df, students_df, how="left", on="school_name")
merge_df
```

# District Summary

```python
##Calculate the total number of schools

    #create an array of unique school names
unique_school_names = merge_df["school_name"].unique()

    #calculate 'len' to get total schools
school_count = len(unique_school_names)    

print(school_count)
```

```python
#Calculate the total number of students
Student_ID_count = merge_df["Student ID"].count()

print(Student_ID_count)
```

```python
#Calculate the total budget

total_budget = schools_df["budget"].sum()
print(total_budget)
```

```python
#Calculate the average math score

avg_math_score = merge_df["math_score"].mean()
print(avg_math_score)

#Calculate the average reading score

avg_reading_score = merge_df["reading_score"].mean()
print(avg_reading_score)
```

```python
#Calculate the overall passing rate (overall average score), 
#i.e. (avg. math score + avg. reading score)/2

num_pass_reading = students_df.loc[students_df["reading_score"] >= 70]["reading_score"].count()
print(num_pass_reading)


```

```python
perc_pass_reading = num_pass_reading/Student_ID_count * 100

print(perc_pass_reading)
```

```python
num_pass_math = students_df.loc[students_df["math_score"] >= 70]["math_score"].count()
print(num_pass_math)


```

```python
perc_pass_math = num_pass_math/Student_ID_count * 100

print(perc_pass_math)
```

```python
#####################FIX THIS!!!!!!#########################
overall_passing_rate = ((perc_pass_reading + perc_pass_math)/2)

print(overall_passing_rate)
```

```python
#Create a dataframe to hold the above results

district_summary = pd.DataFrame({
    
        "Total Schools": [school_count],
        "Total Students": [Student_ID_count],
        "Total Budget": [total_budget],
        "Average Math Score": round(avg_math_score,2),
        "Average Reading Score": round(avg_reading_score,2),
        "Overall Passing Rate": round(overall_passing_rate,2),
        "Total Pass Math": round(num_pass_math,2),
        "% Pass Math": round(perc_pass_math,2),
        "Total Pass Reading": round(num_pass_reading),
        "% Pass Reading": round(perc_pass_reading,2)
    
      })

#store as dif df to change order
dist_sum = district_summary[["Total Schools", 
                             "Total Students",
                             "Total Budget",
                             "Average Math Score",
                             "Average Reading Score",
                             "Overall Passing Rate",
                             "Total Pass Math",
                             "% Pass Math",
                             "Total Pass Reading",
                             "% Pass Reading"]]

#why won't the $ signs or % show up? 
dist_sum.style.format({"Total Budget": "${:,.2f}",
                       "Average Reading Score": "{:.1f}",
                       "Average Math Score": "{:.1f}",
                       "% Pass Reading": "{:.%}",
                       "% Pass Math": "{:.1%}",
                       "Overall Passing Rate": "{:.1%}"})

                    
dist_sum.style
```

```python
#this output has some $ and %

dis_sum_table = pd.DataFrame(district_summary)
dis_sum_table["Total Schools"]
dis_sum_table["Total Students"] = dis_sum_table["Total Students"].map("{:,}".format)
dis_sum_table["Total Budget"] = dis_sum_table["Total Budget"].map("${:,.2f}".format)
dis_sum_table["Overall Passing Rate"] = dis_sum_table["Overall Passing Rate"].map("{:,.2f}%".format)
dis_sum_table["% Pass Math"] = dis_sum_table["% Pass Math"].map("{:,.2f}%".format)
dis_sum_table["% Pass Reading"] = dis_sum_table["% Pass Reading"].map("{:,.2f}%".format)

dis_sum_table
```
